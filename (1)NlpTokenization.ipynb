```python
#corpus for tokenization
corpus = "Amid the rain-drenched streets of Lumina—a city cloaked in mystery!—people hurry beneath flickering streetlights,faces hidden under umbrellas. Shadows dance across walls as the hum of midnight traffic mingles with jazz melodies drifting from hidden bars… In the heart of the city, a forgotten clock tower strikes twelve, its chimes echoing through narrow alleyways like whispers of a lost era. The scent of damp earth and lingering smoke fills the air, blending with murmurs of secrets shared, promises broken, and dreams… forsaken."
```

```python
import nltk
# Download the 'punkt' resource
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize #Splits text into sentences.
sentences = sent_tokenize(corpus)
sentences
```

```python
from nltk.tokenize import word_tokenize #Splits text into individual words or tokens, treating punctuation as separate tokens.
words = word_tokenize(corpus)
words
```

```python
from nltk.tokenize import wordpunct_tokenize #Similar to word_tokenize, but splits at all punctuation marks.
words = wordpunct_tokenize(corpus)
words
```

```python
from nltk.tokenize import TreebankWordTokenizer #Similar to word_tokenize, but splits at all punctuation marks.
tokenizer = TreebankWordTokenizer()
words = tokenizer.tokenize(corpus)
words
```
